```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Importing movie dataset
df_m = pd.read_csv('zeemovies.csv',engine='python', sep='::',names=['MovieID','Title','Genres,,'],encoding='latin-1')
```

```python
df_m.head()
```

```python
df_m.shape
```

```python
#Importing rating dataset
df_r = pd.read_csv('zeeratings.csv',engine='python',sep='::',names=['UserID','MovieID','Rating','Timestamp'],encoding='latin-1')
```

```python
df_r.head()
```

```python
df_r.shape
```

```python
#Importing users dataset
df_u = pd.read_csv('zeeusers.csv',sep='::',names=['UserID','Gender','Age','Occupation','Zip-code'],encoding='latin-1')
```

```python
df_u.head()
```

```python
df_u.shape
```

<font color = 'Black'> <b>Merging all the datasets into one</b>

```python
df_merged1 = df_m.merge(df_r,how='outer')
```

```python
df_merged1.head()
```

```python
df_merged2 = df_u.merge(df_r,how="inner")
```

```python
df_merged2.head()
```

```python
df_merged3 = df_merged1.merge(df_merged2,how="inner")
```

```python
df_merged3.head()
```

```python
df_merged3.shape
```

```python
df_merged3.sort_values(by=['UserID'], ascending=True)
```
